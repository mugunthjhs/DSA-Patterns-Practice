```python
class Solution:
    def hasCycle(self,head):
        slow = head
        fast = head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True
        return False

n1 = Node(1)
n2 = Node(2)
n3 = Node(3)
n4 = Node(4)

n1.next = n2
n2.next = n3
n3.next = n4
n4.next = n2
sol = Solution()
print(sol.hasCycle(n1))
```

```python
n = 19

def happy_number(n):
    
    def next_number(n):
        total = 0
        while n > 0:
            digit = n % 10
            total += digit * digit
            n //= 10
        return total

    slow = n
    fast = next_number(n)

    while fast != 1 and slow != fast:
        slow = next_number(slow)
        fast = next_number(next_number(fast))

    return fast == 1
    

print(happy_number(n))
```

```python
nums = [1,3,4,2,2]

def find_duplicate(nums):
    slow = nums[0]
    fast = nums[0]

    while True:
        slow = nums[slow]
        fast = nums[nums[fast]]
        if slow == fast:
            break
    
    slow = nums[0]
    while slow != fast:
        slow = nums[slow]
        fast = nums[fast]

    
    return slow



print(find_duplicate(nums))
```

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None 

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_node_at_beginning(self,data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        
        temp = self.head
        new_node.next = temp
        self.head = new_node

    def print_all(self):
        temp = self.head
        while temp:
            print(temp.data,end="->")
            temp = temp.next
        print("None")

    def remove_nth_node_from_end(self,index):
        dummy = Node(0)
        dummy.next = self.head

        fast = dummy
        slow = dummy

        for _ in range(index+1):
            fast = fast.next
        
        while fast:
            fast = fast.next
            slow = slow.next
        
        slow.next = slow.next.next

        self.head = dummy.next

    def middle_of_ll(self):
        slow = self.head
        fast = self.head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        self.head = slow
        return 

    def removeElements(self, val):
        prev = 0
        temp = self.head
        
        while temp:
            if temp == val:
                prev.next = temp.next 
            else:
                prev = temp
                temp = temp.next

        return













ll = LinkedList()
ll.insert_node_at_beginning(10)
ll.insert_node_at_beginning(20)
ll.insert_node_at_beginning(30)
ll.insert_node_at_beginning(44)
ll.insert_node_at_beginning(24)
ll.print_all()
ll.removeElements(20)
ll.print_all()
# ll.deleteMiddle()
# ll.print_all()
# ll.insert_node_at_beginning(56)
# ll.deleteMiddle()
# ll.print_all()

# ll.print_all()
# ll.middle_of_ll()
# ll.print_all()

# ll.remove_nth_node_from_end(2)
# ll.middle_of_ll()
# ll.print_all()
        
```
