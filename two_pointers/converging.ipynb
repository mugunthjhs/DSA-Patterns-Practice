```python
def three_sum(nums,target=0):
    nums.sort()
    result = []

    for i in range(len(nums)-2):
        
        if i > 0 and nums[i] == nums[i-1]:
            continue

        left = i + 1
        right = len(nums) - 1

        while left < right:
            s = nums[i] + nums[left] + nums[right]

            if s == target:
                result.append([nums[i],nums[left],nums[right]])
                left += 1
                right -= 1

                while left < right and nums[left] == nums[left - 1]:
                    left += 1
                while left < right and nums[right] == nums[right + 1]:
                    right -= 1
            
            elif s < target:
                left += 1
            else:
                right -= 1

    return result
```

```python
def three_sum_closet(nums,target):
    nums.sort()
    closest = nums[0] + nums[1] + nums[2]

    for i in range(len(nums) - 2):
        if i > 0 and nums[i] == nums[i-1]:
            continue
        left = i+1
        right = len(nums)-1

        while left < right:
            s = nums[i] + nums[left] + nums[right]
            
            if abs(target - s) < abs(target - closest):
                closest = s

            if s < target:
                left += 1
            elif s > target:
                right -= 1
            else:
                return s
    return closest

```

```python
def four_sum(nums,target):
    nums.sort()
    result = []
    n = len(nums)
    
    for i in range(n-3):
        if i>0 and nums[i] == nums[i-1]:
            continue   
        for j in range(i+1,n-2):
            if j>i+1 and nums[j] == nums[j-1]:
                continue
            left = j + 1
            right = n - 1
            while left < right:
                s = nums[i] + nums[j] + nums[left] + nums[right]
                if s == target:
                    result.append([nums[i], nums[j], nums[left], nums[right]])
                    left += 1
                    right -= 1

                    while left < right and nums[left] == nums[left-1]:
                        left += 1
                    while left < right and nums[right] == nums[right+1]:
                        right -= 1
                
                elif s < target:
                    left += 1
                else:
                    right -= 1
    return result
```

```python
people = [3,2,2,1]
limit = 3

def boats_to_save(people,limit):
    people.sort()
    left = 0
    right = len(people) - 1
    boats = 0

    while left <= right:
        if people[left] + people[right] <= limit:
            left += 1

        right -= 1
        boats += 1

    return boats



print(boats_to_save(people,limit))
```

```python
def sorted_square(nums):
    n = len(nums)
    result = [0] * n

    left = 0
    right = n - 1
    pos = n - 1

    while left <= right:
        if abs(nums[left]) > abs(nums[right]):
            result[pos] = nums[left] ** 2
            left += 1
        else:
            result[pos] = nums[right] ** 2
            right -= 1
        pos -= 1
    return result



nums = [-4,-1,0,3,10]
print(sorted_square(nums))
```

```python
nums = [-2,0,1,3]
target = 2

def threeSumSmaller(nums, target):

    nums.sort()
    n = len(nums)
    count = 0

    print("Sorted nums:", nums)
    print("Target:", target)
    print()

    for i in range(n - 2):

        print("---- Fixing i =", i, "value =", nums[i], "----")

        left = i + 1
        right = n - 1

        while left < right:

            print("Current pointers -> i:", i, "left:", left, "right:", right)

            s = nums[i] + nums[left] + nums[right]

            print(
                f"Checking values {nums[i]} + {nums[left]} + {nums[right]} = {s}"
            )

            if s < target:

                valid_count = right - left
                count += valid_count

                print("Sum is smaller than target")
                print("Target =", target)
                print("So this triplet works")

                print(
                    "Because the array is sorted, all elements between left and right also work"
                )

                print(
                    f"So we add (right - left) = ({right} - {left}) = {valid_count} triplets"
                )

                print("Updated total count =", count)

                left += 1
                print("Move left pointer to explore next possibility")
                print()

            else:

                print("Sum is too big or equal to target")
                print("So we decrease the sum by moving right pointer")

                right -= 1
                print("New right pointer =", right)
                print()

    print("Final count =", count)
    return count


threeSumSmaller(nums, target)
```
